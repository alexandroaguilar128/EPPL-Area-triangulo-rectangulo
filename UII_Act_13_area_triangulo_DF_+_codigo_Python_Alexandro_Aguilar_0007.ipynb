## Formula para calcular el area

```python
# Calcular el area de un triangulo rectangulo
# Variable b, h, A
# Formula A = b * h /2
# Capturar el valor de A
b = float(input("Ingresa el valor de la base: "))
h = float(input("Ingresa el valor de la altura: "))
A = b * h / 2
print("El area del triangulo es:", A)
print("Alexandro Aguilar NL = 1")
```
